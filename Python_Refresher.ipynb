```python
# Q1:Program to calculate seconds in given number of days
# Ask user to enter number of days
days = int(input("Enter the number of days: "))
# Convert days to seconds (1 day = 24 hours, 1 hour = 3600 seconds)
seconds = days * 24 * 60 * 60
# Print result
print(days,"days have", seconds, "seconds")
```

```python
# Q2:Program to calculate volume of sphere
import math # import math module to use pi
# ask user to enter radius
r= float(input("Enter radius of sphere: "))
# calculate volume using formula (4/3)*pi*r^3
volume = (4/3) * math.pi*(r**3)
# display results
print("The volume of sphere with radius", r , "is" , round(volume,2))
```

```python
# Q3:Program to compute area and parameter of a square using functions
# function to calculate area
def square_area(side):
    return side * side
# function to calculate perimeter
def square_perimeter(side):
    return 4*side
#ask user to enter length
s = float(input("Enter length of square: "))
# call functions
area = square_area(s)
perimeter = square_perimeter(s)
# display results
print("The area of the square is:", area)
print("The perimeter of the square is:", perimeter)
```

```python
# Q4:Program to check if a character is uppercase or lowercase using a function
# function to check case
def check_case(ch):
    if ch.isupper():
        return "Uppercase"
    elif ch.islower():
        return "Lowercase"
    else:
        return "Not a letter"
#ask user for a character
char = input("Enter a character: ")
#call the function
result = check_case(char)
#display result
print("The character you entered is:",result)
```

```python
# Q5: Program to follow given pseudocode
# set initial values
x = 0
y = 20
# repeat until y is less than 6
while y >= 6:
    y = y-4  #subtract 4 from y
    x = x+(2 / y) #add 2/y to x
# display final value of x
print("Final value of x is:",x)
```

```python
# Q6: Program to input 5 values into a list and calculate their average
numbers = [] #empty list for storing user inputs
# loop 5 times to collect numbers
for i in range(5):
    value = float(input("Enter number " + str(i+1) + ": "))
    numbers.append(value)
#calculate average
average = sum(numbers) / len(numbers)
#display results
print("The numbers you entered are:", numbers)
print("The average is:",average)
```
